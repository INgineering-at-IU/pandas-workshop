```python
# cleaning data
df.loc[0, 'Global_Sales'] = None

df.isnull()
```

```python
#fill null values

df.fillna(value=0, inplace=True)
df
```

```python
#Changing data types to string

df['Year'] = df['Year'].astype(str)
df.dtypes
```

```python
#Changing data types to numeric

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df.dtypes
```

```python
#Creating new Columns

df['NA_Sales_Percentage'] = df['NA_Sales'] / df['Global_Sales']
df['Years_Since_Release'] = 2025 - df['Year']
df
```

```python
#Replacing Year
df['Year'] = df['Year']+5
df
```

```python
# resetting year
df['Year'] = df['Year']-5
```

```python
# Filtering Data
df_twenty_first_century_games = df[df['Year'] >= 2000]
df_twenty_first_century_games
```

```python
# AND (&) operator

df_wii_sports_games = df[(df['Platform']=='Wii') & (df['Genre'] == 'Sports')]
df_wii_sports_games
```

```python
df[(df['Platform']=='Wii') & (df['Genre'] == 'Sports')]
```

```python
# OR (|) operator
df_wii_or_pokemon_games = df[
    (df['Platform'] == 'Wii') | (df['Name'].str.contains('Pokemon', na=False))
]
df_wii_or_pokemon_games
```

```python
#groupby 

df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=True)
```

```python
#descending order

df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)
```

```python
#Get top 5 best selling consoles
df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head()
```
